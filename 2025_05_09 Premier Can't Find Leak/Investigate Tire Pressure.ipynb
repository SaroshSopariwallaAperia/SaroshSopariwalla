# Load Operational Pressure Data for Right Steer

```python
schema = account_schema_mapping['MTS']
query = f"""
SELECT DISTINCT 
    fleet_meta_data.code,
    fleet_meta_data.fleet_name,
    vehicle_meta_data.vehicle_id,
    vehicle_meta_data.fleet_vehicle_id,
    meta_data.side || meta_data.axle || meta_data.position as tire,
    meta_data.type,
    meta_data.set_point,
    sensor_data.pressure_date,
    sensor_data.pressure,
    sensor_data.latitude,
    sensor_data.longitude,
    sensor_data.speed,
    sensor_data.temperature
FROM
    {schema}.fleet_meta_data
    INNER JOIN {schema}.vehicle_meta_data ON fleet_meta_data.fleet_id = vehicle_meta_data.fleet_id
    INNER JOIN {schema}.meta_data ON vehicle_meta_data.vehicle_id = meta_data.vehicle_id
    INNER JOIN {schema}.sensor_data ON meta_data.id = sensor_data.meta_data_id
WHERE
    vehicle_meta_data.vehicle_id = 11352
    AND sensor_data.pressure_date >= '2025-05-08'
    AND meta_data.side || meta_data.axle || meta_data.position = 'R3O'
ORDER BY tire, pressure_date ASC
"""
df = pd.read_sql(query,engine)
```

```python
df.to_excel(f"{df['fleet_name'].iloc[0]} {df['fleet_vehicle_id'].iloc[0]} {df['tire'].iloc[0]} Sensor Data.xlsx",index = False)
```
